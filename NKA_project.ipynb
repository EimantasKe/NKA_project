```python
doc = nlp(news)
entities = ['ORG', 'LOC', 'PERSON']
people = []
locations = []
orgs = []
entity_list = []

for entity in doc.ents:
  entity_list.append(entity.text)
  if entity.label_ in entities:
    print(entity.label_, ' | ', entity.text)
    if entity.label_ in "ORG":
      orgs.append(entity.text)
    if entity.label_ in "LOC":
      locations.append(entity.text)
    if entity.label_ in "PERSON":
      people.append(entity.text)
```

```python
colors = {'ORG':'linear-gradient(180deg, green, yellow)', 'LOC':'radial-gradient(yellow, red)', 'PERSON':'linear-gradient(90deg, #aa9cfc, #fc9ce7)'}
options = {'ents': entities, 'colors':colors}
displacy.render(doc, style='ent', jupyter=True, options=options)
```

```python
labels = ['People', 'Locations', 'Organizations']
values = [len(people), len(locations), len(orgs)]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(labels, values)
fig.suptitle("Count of entities")
plt.savefig("fig1.png", bbox_inches = "tight")
plt.show()
```

```python
labels = [ 'People', 'Locations', 'Organizations']
values = [ len(people), len(locations), len(orgs)]

fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
fig1.suptitle("Entity percentage split")
plt.savefig("fig2.png", bbox_inches = "tight")
plt.show()
```

```python
labels = ['Other entities', 'People', 'Locations', 'Organizations']
values = [len(entity_list) - (len(people) + len(locations) + len(orgs)), len(people), len(locations), len(orgs)]
explode = (0, 0.1, 0.1, 0.1)

fig1, ax1 = plt.subplots()
ax1.pie(values, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
fig1.suptitle("Target entities v. all others")
plt.savefig("fig3.png", bbox_inches = "tight")
plt.show()
```
